# Training

```python
import numpy as np


diabetes_data = np.load('processed_diabetes_data.npy', allow_pickle=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.DataFrame(diabetes_data)
df.drop(6, axis=1, inplace=True)
df.insert(0, '', 1)


X = df.to_numpy()
y = diabetes_data[:, 6]

X_train, X_test, y_train, y_test = train_test_split(X, y)

```

Train Logistic Regression Model

```python
import LogisticRegression_GD as LR

Logistic = LR.logistic_regression('logistic_regression')

parameter_vector = np.array([
    [1, 1, 1, 1, 1, 1, 1, 1, 1], #first term is bias
])

learning_rate = 0.1

epochs = 3000

X_train = X_train.astype(np.float64)
y_train = y_train.astype(np.int64)

LR_params, LR_param_history = Logistic.gradient_descent(parameter_vector, X_train, y_train, epochs, learning_rate)
```

# Performance Measures

Predict on the test set

```python
m, n = X_test.shape

prediction_threshold = 0.5 

X_test= X_test.astype(np.float64)

p_hat = Logistic.probability(LR_params, X_test)
y_hat = Logistic.model_prediction(m, prediction_threshold, p_hat)
```

Match Shapes for Performance Measures

```python
y_hat = np.squeeze(y_hat.T) #change 2 dimmnesional array to one dimmensional array
```

Precision, Accuracy, Recall

```python
#accuracy - correct predictions ratio
correct_predictions = y_test == y_hat #boolean mask

accuracy_sum = np.zeros((m,)) 
accuracy_sum[correct_predictions] = 1 
TP_plus_TN = np.sum(accuracy_sum)

accuracy = TP_plus_TN / m

print(f'Accuracy: {accuracy}')


#precision -- accuracy of positive predictions
positive_predictions = y_hat == 1  #boolean mask
pos_pred_sum = np.zeros(m,)
pos_pred_sum[positive_predictions] = 1

TP_FP = np.sum(pos_pred_sum)

TP_mask = (y_hat == 1) & (y_test == 1)
TP_sum = np.zeros(m,)
TP_sum[TP_mask] = 1
TP = np.sum(TP_sum)

precision = TP / TP_FP

print(f'Precision: {precision}')


#Recall - ration of pos instances correctly classified
positive_instances = y_hat == 1 #boolean mask
pos_inst_sum = np.zeros(m,)
pos_inst_sum[positive_instances] = 1

TP_FN = np.sum(pos_inst_sum)

recall = TP / TP_FN

print(f'Recall: {recall}')
```

That's Not Bad!

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

y_test = y_test.astype(np.int64)
y_hat = y_hat.astype(np.float64)

print(confusion_matrix(y_test, y_hat))
```
